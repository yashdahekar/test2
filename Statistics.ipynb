# question 1 

- Six Sigma is a methodology and statistical system aimed at achieving nearly flawless process outputs by minimizing variability. In simpler terms, it focuses on striving for minimal defects or errors.

- Sigma (σ): This symbol denotes the standard deviation, a statistical measure indicating the spread of data points from the average in a normal distribution (bell curve).

- Six Sigma level: A process reaches "Six Sigma" status when it produces only 3.4 defects per million opportunities (DPMO). This signifies a process with minimal variation, resulting in exceptionally high quality.

- Example: Reducing Call Center Wait Times

A call center faces challenges with high call volumes, leading to extended wait times and dissatisfied customers. Implementing Six Sigma can enhance this process.

  * Define: Establish a target for average call wait time (e.g., less than 5 minutes).
  * Measure: Monitor call wait times and identify peak call periods.
  * Analyze: Utilize statistical tools to identify factors influencing wait times (e.g., staffing levels, call complexity, agent training).
  * Improve: Introduce changes to minimize variations (e.g., increase agent numbers during peak hours, categorize calls for quicker routing, provide additional training for complex calls).
  * Control: Continuously monitor wait times and make necessary adjustments.

By applying Six Sigma principles, the call center can:

  * Decrease customer wait times, enhancing satisfaction.
  * Improve agent efficiency by streamlining call handling processes.
  * Potentially reduce staffing costs through optimized scheduling.

- Discrete or categorical data does not adhere to a log-normal or Gaussian distribution, as these distributions are continuous probability distributions.

- Instead, discrete data often conforms to distributions such as the Beta distribution, Gamma distribution, Poisson distribution, or Uniform distribution, among others.

- For instance, data concerning the number of accidents typically follows a Poisson distribution.

- Similarly, the lifetimes of products or machines tend to follow a Weibull distribution.

- The concept of a 5 number summary is a way to describe a distribution using 5 numbers.

-  * Minimum Value: The smallest observation in the dataset  and the first number when it is sorted in ascending order.
   * First Quartile (Q1): The value below which 25% of the data falls.first quartile = ((n + 1)/4)th term
   * Median (Q2): The middle value that splits the dataset into two halves. Median = (n + 1)/2th term
   * Third Quartile (Q3): The value below which 75% of the data falls.Third quartile = (3(n + 1)/4)th term
   * Maximum Value: The largest observation in the dataset , and the last number when it is sorted in ascending order.

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns


iris = load_iris()


data= pd.DataFrame(iris.data)
data.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
```

```python
data
```

```python
# Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='sepal_length', y='sepal_width')
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
# Pairplot
sns.pairplot(data)
plt.suptitle('Pairplot of Iris Dataset', y=1.02)
plt.show()
```

```python
# Heatmap
correlation_matrix = data.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Iris Dataset')
plt.show()
```
